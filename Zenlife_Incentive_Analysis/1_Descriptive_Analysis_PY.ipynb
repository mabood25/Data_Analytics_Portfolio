```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfr = pd.read_excel(r'Zenlife original.xlsx')
pd.set_option('display.max_rows', 114)
```

```python
dfr.head(5)
```

```python
dfr.columns = ['Week', 'Costa Rica', 'Panama', 'Colombia', 'Ecuador']
dfr.drop(index=[0, 1, 2], inplace=True)
dfr.reset_index(inplace=True)
dfr.drop(columns=['index'], inplace=True)
```

```python
dff = dfr.copy()
col_to_x = ['Costa Rica', 'Panama', 'Colombia', 'Ecuador']
dff[col_to_x] = (dff[col_to_x] * 1000).astype(float)
dff.to_excel('Zenlife PowerBI.xlsx', index=False)
```

```python
dfr.set_index('Week', inplace=True)
dfr = dfr.astype(float)
```

```python
description = dfr.describe()
sum_row = dfr.sum()
description.loc['sum'] = sum_row
print(description)
```
